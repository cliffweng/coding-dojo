## Probability Distributions
Common statistical distributions https://www.datacamp.com/community/tutorials/probability-distributions-python

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Uniform Distribution

```python
from scipy.stats import uniform
data_uniform = uniform.rvs(size=10000, loc = 10, scale=20) # generate a ndarray of 10,000 evenly distributed between 10 and 30 (10+20)
```

```python
ax = sns.displot(data_uniform,bins=100, kde=True, height=4, aspect=2)
ax.set(xlabel='Uniform Distribution ', ylabel='Frequency')
```

## Normal Distribution

```python
from scipy.stats import norm
data_normal = norm.rvs(size=10000,loc=20,scale=10) # loc : mean of the distribution. scale: standard deviation
```

```python
ax = sns.displot(data_normal,bins=100, kde=True, height=4, aspect=2)
ax.set(xlabel='Normal Distribution ', ylabel='Frequency')
```

## Gamma Distribution
The gamma distribution is a two-parameter family of continuous probability distributions. While it is used rarely in its raw form but other popularly used distributions like exponential, chi-squared, erlang distributions are special cases of the gamma distribution.

```python
from scipy.stats import gamma
```

```python
ax = sns.displot(gamma.rvs(a=10, size=1000,loc=10),bins=100, kde=True, height=2, aspect=4)
ax.set(xlabel='Gamma Distribution (10)', ylabel='Frequency')
ax = sns.displot(gamma.rvs(a=5, size=1000,loc=10),bins=100, kde=True, height=2, aspect=4)
ax.set(xlabel='Gamma Distribution (5)', ylabel='Frequency')
ax = sns.displot(gamma.rvs(a=2, size=1000,loc=10),bins=100, kde=True, height=2, aspect=4)
ax.set(xlabel='Gamma Distribution (2)', ylabel='Frequency')
ax = sns.displot(gamma.rvs(a=1, size=1000,loc=10),bins=100, kde=True, height=2, aspect=4)
ax.set(xlabel='Gamma Distribution (1)', ylabel='Frequency')
```

## Poisson Distribution
Poisson random variable is typically used to model the number of times an event happened in a time interval. For example, the number of users visited on a website in an interval can be thought of a Poisson process. 

```python
from scipy.stats import poisson
```

```python
ax = sns.displot(poisson.rvs(mu=20, size=1000), bins=20, kde=True, height=3, aspect=2)
ax.set(xlabel='Poisson Distribution ', ylabel='Frequency')
ax = sns.displot(poisson.rvs(mu=40, size=1000), bins=20, kde=True, height=3, aspect=2)
ax = sns.displot(poisson.rvs(mu=100, size=1000), bins=20, kde=True, height=3, aspect=2)
```

## Binomial Distribution

## Bernoulli Distribution
A Bernoulli distribution has only two possible outcomes, namely 1 (success) and 0 (failure), and a single trial, for example, a coin toss. So the random variable X which has a Bernoulli distribution can take value 1 with the probability of success, p, and the value 0 with the probability of failure, q or 1p.

```python
from scipy.stats import bernoulli
ax = sns.displot(bernoulli.rvs(p=0.8, size=1000), kde=False, height=2, aspect=4)
ax.set(xlabel='Bernoulli Distribution (p=0.8) ', ylabel='Frequency')
ax = sns.displot(bernoulli.rvs(p=0.5, size=1000), kde=False, height=2, aspect=4)
ax.set(xlabel='Bernoulli Distribution (p=0.5) ', ylabel='Frequency')
```
